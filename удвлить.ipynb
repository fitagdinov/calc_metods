```python
Tableau = []
Tableau.append(0)

Tableau1 = [[[0,0]],
               [1]]

Tableau2 = [[[0,0,0],
             [1/2,1/2,0]],
                  [0,1]]

Tableau3 = [[[0,0,0,0],
             [1/2,1/2,0,0],
             [1,-1,2,0]],
               [1/6,2/3,1/6]]

Tableau4 = [[[0,0,0,0,0],
             [1/2,1/2,0,0,0],
             [1/2,0,1/2,0,0],
             [1,0,0,0,1]],
               [1/6,2/3,1/3,1/6]]

Tableau.append(Tableau1)
Tableau.append(Tableau2)
Tableau.append(Tableau3)
Tableau.append(Tableau4)
```

```python
import numpy as np
import scipy.linalg as la

M = [[0,1],[-100,-101]]
def f(t,x):
    return np.dot(M,x)

eigein = la.eig(M)
print("Eigeinvalues ", eigein[0])
print("Eigeinvectors ", eigein[1])
```

```python
def solve(tab,p,s,x):
    for i in range(1,n):
        k = np.zeros((p,s))
        for j in range(p):
            dx = np.zeros(s)
            
            for l in range(p):
                dx += np.dot(dt*tab[0][j][1+l],k[l])
                
            k[j] = f(t[i-1]+dt*tab[0][j][0],x[i-1] + dx)

        t[i] = t[i-1] + dt
        
        dx = np.zeros(s)
        for l in range(p):
            dx += np.dot(tab[1][l],k[l])
            
        x[i] = x[i-1] + dx*dt
```

Найдём области устойчивости<br><br>
    Функции устойчивости для явных методов 1-4 порядков я нашёл в задаче 7.2<br><br>
    1) $R_{1}(z) = 1 + z$<br>
    2) $R_{2}(z) = 1 + z + \frac{z^2}{2!}$<br>
    3) $R_{3}(z) = 1 + z + \frac{z^2}{2!} + \frac{z^3}{3!}$<br>
    4) $R_{4}(z) = 1 + z + \frac{z^2}{2!} + \frac{z^3}{3!} + \frac{z^4}{4!}$<br>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4,3,0.05)
y = np.arange(-4,4,0.05)
xx,yy = np.meshgrid(x,y)
zz = xx + 1j*yy     

def R(z,p):
    if p == 1:
        return 1+z
    if p == 2:
        return 1+z+(z**2)/2
    if p == 3:
        return  1+z+(z**2)/2+(z**3)/6
    if p == 4:
        return 1+z+(z**2)/2+(z**3)/6+(z**4)/24


fig, ax = plt.subplots(4, figsize = (12,24))

z1 = abs(R(zz,1))
z2 = abs(R(zz,2))
z3 = abs(R(zz,3))
z4 = abs(R(zz,4))

cf1 = ax[0].pcolor(x,y,z1,vmin=np.amin(z1),vmax=1)
ax[0].contour(x,y,z1,[1],colors='black')

cf2 = ax[1].pcolor(x,y,z2,vmin=np.amin(z2),vmax=1)
ax[1].contour(x,y,z2,[1],colors='black')

cf3 = ax[2].pcolor(x,y,z3,vmin=np.amin(z3),vmax=1)
ax[2].contour(x,y,z3,[1],colors='black')

cf4 = ax[3].pcolor(x,y,z4,vmin=np.amin(z4),vmax=1)
ax[3].contour(x,y,z4,[1],colors='black')


cbar1 = fig.colorbar(cf1,ax=ax[0])
cbar2 = fig.colorbar(cf2,ax=ax[1])
cbar3 = fig.colorbar(cf3,ax=ax[2])
cbar4 = fig.colorbar(cf4,ax=ax[3])

ax[0].grid();ax[1].grid();ax[2].grid();ax[3].grid()
```
